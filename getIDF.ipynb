```python
from konlpy.tag import Okt
from collections import OrderedDict
import numpy as np
import json
from math import log10
import pickle
import pandas as pd
import re

file = open("vsm_raw_fix.pickle","rb")
c = pickle.load(file)
list_word = list(c.keys())
file.close
            
print(len(list_word))

tf = np.zeros((len(list_word),10),dtype = float)

count = 0

for i in range(10):
    rate = i+1
    print(i," docu")
    count = 0
    file ='data\\rating{}_docs.json'.format(i+1)
    with open(file,encoding = "utf-8") as data_file:
        da = json.load(data_file, object_pairs_hook=OrderedDict)
        for m in range(len(da)):
            for j in range(len(da[m])):
                text = da[m][j]
                if(text in list_word):
                    tf[list_word.index(text),i] += 1
                    count+=1
                    if count%100000 == 0:
                        print(count)
        max = np.max(tf[:,i])
        for k in range(len(list_word)):
            tf[k,i] = 0.5 + (0.5 * tf[k,i])/max
                    
                
csv = pd.DataFrame(tf)
csv.to_csv("tf_case2_a.csv",header = False, index = False)
```
