```python
from konlpy.tag import Okt
from collections import OrderedDict
import numpy as np
import json
from math import log10
import pickle

def f(t, d):
    return d.count(t)

def tf(t, d):
    return 0.5 + 0.5*f(t,d)/max([f(w,d) for w in d])

def idf(t, D):
    # D is documents == document list
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

def tfidf(t, d, D):
    return tf(t,d)*idf(t, D)


# with open('vsm_over100.txt','r',encoding="utf8") as file:
#     vsm = json.load(file)
# print(len(vsm))

file = open('vsm_raw_fix.pickle',"rb")
c = pickle.load(file)
file.close

list_word = list(c.keys())

count = 0

size = 30000

data = np.zeros((10,size),dtype = float)
score = np.zeros(size*10,dtype = int)

for i in range(10):
    rate = i+1
    file ='data\rating{}_docs.json'.format(i+1)
    with open(file,encoding = "utf-8") as data_file:
        da = json.load(data_file, object_pairs_hook=OrderedDict)
        for m in range(len(da)):
            if(m == size):
                break
            data[i,:] = da[m]

```
