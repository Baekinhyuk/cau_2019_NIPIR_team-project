```python
from konlpy.tag import Okt
from collections import OrderedDict
import numpy as np
import json
from math import log10
import pickle
import pandas as pd
import re

def f(t, d):
    return d.count(t)

def tf(t, d):
    return 0.5 + 0.5*f(t,d)/max([f(w,d) for w in d])

def idf(t, D):
    # D is documents == document list
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

def tfidf(t, d, D):
    return tf(t,d)*idf(t, D)


# with open('vsm_over100.txt','r',encoding="utf8") as file:
#     vsm = json.load(file)
# print(len(vsm))

file = open('vsm_newnew.pickle',"rb")
list_word = pickle.load(file)
file.close

tf = np.zeros((len(list_word),10),dtype = float)

count = 0
print(len(list_word))

print(list_word)

for i in range(10):
    rate = i+1
    file ='data\\rating{}_docs.json'.format(i+1)
    with open(file,encoding = "utf-8") as data_file:
        da = json.load(data_file, object_pairs_hook=OrderedDict)
        for m in range(len(da)):
            for j in range(len(da[m])):
                text = re.sub('[-=.#/?:$}!~`"@,0-9)(]', '', da[m][j])
                text = re.sub('[ㅊㅋ]+','ㅊㅋ',text)
                text = re.sub('ㅋ+','ㅋ',text)
                text = re.sub('ㅎ+','ㅎ',text)
                text = re.sub('[ㅡㅡ]+','ㅡㅡ',text)
                text = re.sub('ㅠ+','ㅠ',text)
                text = re.sub('ㅜ+','ㅜ',text)
                text = re.sub('ㅗ+','ㅗ',text)
                text = re.sub('ㄱ+','ㄱ',text)
                
                if(text in list_word):
                    tf[list_word.index(text),i] += 1
        tf[:,i] /= len(da)
                
csv = pd.DataFrame(tf)
csv.to_csv("tf_new.csv",header = False, index = False)
```
