```python
# vsm 줄이기 코드

import pickle
import json
from pprint import pprint as pp

threshold = 50 # 빈도수 체크

file = open('vsm_raw.pickle',"rb")
c = pickle.load(file)
file.close

new_vsm = {}

list_item = list(c.items())

del c['']
list_item.sort(key=lambda x:x[1], reverse = True)
count = 0;
for i in list_item:
    if(i[1] < threshold):
        del c[i[0]]
        
#===============================

list_item = list(c.items())
list_item.sort(key=lambda x:x[1], reverse = True)
for i in list_item:
    if(i[1] > threshold):
        new_vsm[i[0]] = i[1]
        count += 1
    
print('total : ',count)

with open('vsm_raw_fix.pickle', 'wb') as f:
    pickle.dump(new_vsm, f, pickle.HIGHEST_PROTOCOL)
    f.close
print("done")
```

```python
import pickle
import re
import json
from konlpy.tag import Okt
from collections import OrderedDict
import numpy as np

dict = {}

for i in range(10):
    rate = i+1
    file ='data\\rating{}_docs.json'.format(i+1)
    with open(file,encoding = "utf-8") as data_file:
        da = json.load(data_file, object_pairs_hook=OrderedDict)
        for m in range(len(da)):
            for j in range(len(da[m])):
                text = re.sub('[-=.#/?:$}!~`"@,0-9)(]', '', da[m][j])
                text = re.sub('[ㅊㅋ]+','ㅊㅋ',text)
                text = re.sub('ㅋ+','ㅋ',text)
                text = re.sub('ㅎ+','ㅎ',text)
                text = re.sub('[ㅡㅡ]+','ㅡㅡ',text)
                text = re.sub('ㅠ+','ㅠ',text)
                text = re.sub('ㅜ+','ㅜ',text)
                text = re.sub('ㅗ+','ㅗ',text)
                text = re.sub('ㄱ+','ㄱ',text)
                if(text in dict):
                    dict[text] = dict[text] +1
                else:
                    dict[text] = 1
                    
with open('vsm_raw.pickle', 'wb') as f:
    pickle.dump(dict, f, pickle.HIGHEST_PROTOCOL)
    f.close
    
print("done")
```
